# Informe sobre MOOCS

# Introducción

El presente informe surge como una síntesis del trabajo realizado para responder a las preguntas acerca de cuál es estado de situación actual de la industria de la educación en línea, particularmente en la rama de creación y oferta de cursos masivos y abiertos on-line, conocidos como MOOCS por sus siglas en inglés. 
Para realizar el trabajo, se nos proveyó de un conjunto de datasets que contenían datos sobre tres plataformas líderes del rubro. 

En el proyecto utilizamos tanto técnicas de ingeniería de datos, para tratar los datasets que nos fueron entregados; como técnicas de visualización de datos. Entre las herramientas utilizadas en estos procesos podemos mencionar a Python como lenguaje de programación. Además, nos servimos de librerías especializadas de este lenguaje, algunas específicas para el tratamiento de datos; otras en la visualización; ya sea, estas últimas, como soporte para la exploración de datos o para la visualización de la información obtenida. Entendemos como librerías a los módulos especializados de un lenguaje de programación que permiten realizar una tarea en particular, o un conjunto de tareas fuertemente ligadas entre sí.

## Metodología 

La metodología que utilizamos para este proyecto consistió entoces en tomar los datos del modelo Edx-Coursera y contrastarlos con los datos de la plataforma Udemy. Es decir, algunas veces contrastamos datos de Coursera vs Udemy, cuando sus campos coincidían y otras veces contrastamos los datos de Edx con los de Udemy. Esto nos permitió, en parte sortear el problema que plateaba 

#### Consideraciones e imputaciones

Un dato que nos pareció interesante extraer fue el esfuerzo que llevaba realizar cada curso. En edx esta información estaba concentrada en la columna 'course_effort' que contiene el tiempo estimado en horas y la duración del curso. Nosotros tomamos la duración del curso y la multiplicamos por el promedio de horas de cada celda. Es otras palabras, cada celda contiene un rango de horas estimado, nosotros tomamos el punto medio de ese rango y lo multiplicamos para obtener el esfuerzo estimado en horas.

Por el lado de Udemy contábamos con la duración en horas del contenido, es decir, con el tiempo total de los videos. Para calcular el esfuerzo total en esta plataforma multiplicamos por 4 este número, de manera de contemplar el esfuerzo que conlleva realizar otras tareas además de ver el curso. 

Con todo esto ya contamos con un campo de esfuerzo total para ambos modelos

Otro dato que nos pareció interesante estimar fue el de la fecha de comienzo de los cursos en el modelo edx-coursera. Sin embargo, no contábamos con fechas de comienzo de cursos para estas plataformas. No obstante, el datasets de reviews de Coursera 
contiene fechas de los reviews. Por lo tanto, tomamos la fecha del primer review de cada curso, esta nos da un indicio de cuando el curso fue creado. Entendemos que no es la fecha exacta de publicación del curso pero si constituye una buena aproximación. 

En cuanto a algunos datos extras que conseguimos, estos surgen de utilizar algoritmos de web scraping para conectarse remotamente a los servidores de estas plataformas y conseguir información pertinente. De esta manera pudimos conseguir un listado de ratings para Udemy y los sitios de cursos de esta plataforma que todavía siguen activos.

# Conclusiones

El recorrido de nuestro trabajo nos ha llevado a las siguientes conclusiones. Lo que en principio notamos como dos posibles modelos de negocio fue haciéndose cada vez más claro. A continuación mencionamos lo hallado en el análisis y exponemos nuestras conclusiones para las dimensiones analizadas.

Primero, lo que pudimos encontrar fue que las tres plataformas crean gran cantidad de contenido para principiantes, esto se convierte en una de las bases del negocio, pero también contrario a lo que pensamos en principio, hay una gran cantidad de ofertas de otros niveles. Por lo tanto la recomendación en cuanto a este punto será: crear cursos de todos los niveles, poniéndo especial énfasis en el nivel de principiante, sin descuidar el contenido para los demás niveles, de esta manera se logra diversificar la oferta, atraer a nuevos clientes independientemente de su nivel de conocimiento y retener a aquellos clientes que dieron sus primeros pasos ofreciéndoles cursos para seguir avanzando en el área de conocimiento que hayan elegido. 

Segundo, los modelos de negocios vistos difieren en esfuerzo que demandan a sus estudiantes. Nuestra hipótesis es que esta dimensión está relacionada con quienes imparten los cursos. En el modelo Edx-Coursera al ser mayormente cursos impartidos por universidades, estas ya cuentan con un gran conocimiento acumulado acerca de como organizar y transmitir el conocimiento. Por lo tanto es de esperar que puedan planificar mejor cursos para el mediano plazo, e incluso para el largo plazo. Además al contar con equipos de personas, la producción de material es menos demandante que en el caso de ser realizada por particulares.
En cuanto a este punto recomendamos que si la empresa quiere crear contenido de mediano y largo plazó deberá optar por un modelo de negocio como el de Edx-Coursera, mientras que si quiere crear contenido de un período corto a mediano deberá buscar un modelo similar al de Udemy.

Tercero: Notamos en Udemy un crecimiento de la oferta muy brusco, seguido por un declive bastante marcado en este crecimiento, esto debido a que cada año presentó una cantidad de cursos menor que el anterior a partir de 2016. Este hecho sumado a la cantidad de cursos que quedaron fuera de línea habla de una oferta no tan estables, quizás podríamos hablar de una rotación de cursos. Si bien la flexibilidad de que los particulares puedan ofrecer cursos puede el fundamento de este negocio, esto también puede ser una debilidad para una empresa que recién se inicie en este negocio, debido que al no brindar suficiente estabilidad en la oferta esto puede desalentar a los potenciales nuevos clientes. Por el otro lado, la oferta de Coursera se ve más estable a lo largo del tiempo. En este punto nuestra recomendación es crear una base de cursos inpartidos por instituciones académicas y organizaciones privadas, dado que la oferta más estable tiende a atraer y a retener a los clientes. Si se decide una estrategia como la de Udemy debería contarse con una fuerte campaña para atraer creadores de una manera rápida y eficaz, de manera de poder soportar la rotación de oferentes.

Cuarto: Los cursos más masivos tienen como condición necesaria un alto rating. Por lo tanto recomendamos no solo tener como meta crear contenido de calidad, sino también crear por un lado, mecanismos para que los alumnos puedan votar el rating de sus cursos y por otro lado crear canales que comuniquen estos ratings a otro alumnos o a potenciales nuevos clientes.

Quinto: Ya observamos que las plataformas tienen una oferta de cursos que explota en los primeros años. A esto debe apuntar la empresa, a crear una gran oferta en los primeros años de vida de la empresa, que funcionarán como base para atraer a nuevas instituciones y organismos. En caso de ir con un negocio como el de Udemy, ya mencionamos más arriba la creación de una campaña para atraer masivamente a los individuos creadores de cursos. 

Por último: Los cursos que presentan mayores subscriptores son aquellos en los rangos más bajos, los cursos con un precio menor o igual que 50 dolares conforman el 54 % de la oferta. Aquí tenemos, entoces, un claro indicio de cuales son los rangos de precios que la empresa debe manejar al comienzo de sus funciones. Una buena estrategia sería salir al mercado con una oferta que tenga  una gran proporción de cursos dentro de estos rangos.
